```python
# Week 1 â€” Load & Inspect Dataset
# Tujuan: Belajar kenalan sama data kita.

# Import library
import pandas as pd
from pathlib import Path

# 1ï¸âƒ£ Set Path File
raw_data_path = Path("../data/raw/sample_dataset.csv")
processed_data_path = Path("../data/processed/data_overview.csv")

# 2ï¸âƒ£ Load Dataset
try:
    df = pd.read_csv(raw_data_path)
    print("âœ… Data berhasil dibaca")
except Exception as e:
    print(f"âŒ Gagal membaca data: {e}")

# 3ï¸âƒ£ Lihat bentuk & info dataset
print("ğŸ“ Bentuk dataset:")
print(df.shape)

print("\nâ„¹ï¸ Info dataset:")
print(df.info())

print("\nğŸ” 5 baris pertama:")
display(df.head())

print("\nğŸ” 5 baris terakhir:")
display(df.tail())

# 4ï¸âƒ£ Cek Missing Values
print("\nâŒ Missing Values tiap kolom:")
display(df.isna().sum())

# 5ï¸âƒ£ Cek duplikat
num_dupes = df.duplicated().sum()
print(f"\nğŸ§® Jumlah duplikat: {num_dupes}")

# 6ï¸âƒ£ Ringkasan / Overview
overview = []

for col in df.columns:
    overview.append({
        "column_name": col,
        "data_type": str(df[col].dtype),
        "missing_values": df[col].isna().sum(),
        "unique_values": df[col].nunique(),
        "sample_value": df[col].dropna().iloc[0] if not df[col].dropna().empty else None
    })

overview_df = pd.DataFrame(overview)

# 7ï¸âƒ£ Simpan overview
processed_data_path.parent.mkdir(parents=True, exist_ok=True)
overview_df.to_csv(processed_data_path, index=False)
print(f"\nğŸ’¾ Overview dataset disimpan di: {processed_data_path}")

# 8ï¸âƒ£ Tampilkan Overview
print("\nğŸ“Š Overview dataset:")
display(overview_df)
```
